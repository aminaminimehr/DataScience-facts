```python
import pytse_client as tse
import pandas as pd

vabemelat = tse.Ticker("وبملت")
vabemalat = vabemelat.history

print(vabemalat)
```

```python
df = pd.DataFrame(vabemalat)
df['return'] = df['close'].pct_change()
```

```python
list_sahm = ["وبملت" , "شیران" , "تاپیکو" , "حتوکا" , "شکربن" , "غاذر" , "فاسمین" , "فملی" , "وتجارت" , "کگل" , "خودرو" , "خپارس"]
list_sabz_ha = []
lis_ghermez_ha = []
for i in list_sahm :
    sahm = tse.Ticker(i)
    sahm = sahm.history
    df = pd.DataFrame(sahm)
    df['return'] = df['close'].pct_change()
    if df['return'][-1] > 0 :
        list_sabz_ha.append(i)
    else:
        lis_ghermez_ha.append(i)
```
